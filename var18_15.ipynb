```python
 # эффективная площадь раскрыва
A1=2.8
η2= 0.86
h= 1050   # длина радиолинии в километрах

Δf= 5000*(10**3)  # полоса пропускания 
T0=300
radius=6370 # радиус земли в километрах
```

Шумовая температура направленной антенны вычисляется по формуле,    
Ткосм=Та
T_cos для каждой волны своя 

```python
# для 3 см
t_cos1=0
t_a_3=t_cos1
print ('Для 3 см '+f'{t_a_3} K')

# для 15 см
t_cos2=8

t_a_15=t_cos2
print('Для 15 см '+f'{t_a_15} K')

# для 30 см
t_cos3=40
t_a_30=t_cos3
print('Для 30 см '+f'{t_a_30} K')

# для 150 см
t_cos4=1000
t_a_150=t_cos4
print('Для 150 см '+f'{t_a_150} K')

# для 300 см
t_cos5=7000
t_a_300=t_cos5
print('Для 300 см '+f'{t_a_300} K')


```

расчитаем длину радиолинии в метрах

```python
rad=6370*(10**3)    # Радиус земли в метрах
hi=1050*(10**3)      # Высота в метрах
```

```python
r=2*((((rad+hi)**2)-((rad)**2))**(1/2)) # длина радиолинии в метрах
print(r)
```

Посчитаем моножитель ослабления W  
где E=50 (мкВ/м определили по таблице)

```python
E=50*(10**(-3))
W=(E*rad)/(3*(10**5))
print ('моножитель ослабления W='+f'{W}')
```
