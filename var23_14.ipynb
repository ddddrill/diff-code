Данные нам величины  
A1= 1,6 м^2  
η2= 0.93  
h= 450 км  
r= 1000 км  
Δf= 90 кГц  
Вид радиосвязи - Фототелеграфия

β – коэффициент различимости по напряженности поля (5,10);  
А2 – эффективная площадь приемной антенны, м2;

W – множитель ослабления;

Rз- радиус земли

```python
A1= 1.6 # эффективная площадь раскрыва
η2= 0.93
h= 450 
r= 1000*(10**3) 
Δf= 90
T0=300
radius=6370
```

### Сначала выполним расчеты для N=1

```python
N_1=1
```

Шумовая температура радиоприемного устройства приближенно равна  
Tшрпу=(N-η2)*T0

```python
Tnrpu_1=(N_1-η2)*T0
```

Шумовая температура ненаправленной (реально – слабонаправленной) антенны вычисляется по формуле,    
где T_le-яркостная температура шумов Земли,T_cos-яркостная температура космических шумов  
T_cos для каждой волны своя 

```python
# для 3 см
t_cos1=0
t_le=300
t_a_3=(1/2)*(t_le*(1-(1-(radius/(radius+h))**2)**(1/2))+t_cos1*(1+(1-(radius/(radius+h))**2)**(1/2)))
print ('Для 3 см '+f'{t_a_3} K')

# для 15 см
t_cos2=8
t_le=300
t_a_15=(1/2)*(t_le*(1-(1-(radius/(radius+h))**2)**(1/2))+t_cos2*(1+(1-(radius/(radius+h))**2)**(1/2)))
print('Для 15 см '+f'{t_a_15} K')

# для 30 см
t_cos3=40
t_le=300
t_a_30=(1/2)*(t_le*(1-(1-(radius/(radius+h))**2)**(1/2))+t_cos3*(1+(1-(radius/(radius+h))**2)**(1/2)))
print('Для 30 см '+f'{t_a_30} K')

# для 150 см
t_cos4=1000
t_le=300
t_a_150=(1/2)*(t_le*(1-(1-(radius/(radius+h))**2)**(1/2))+t_cos4*(1+(1-(radius/(radius+h))**2)**(1/2)))
print('Для 150 см '+f'{t_a_150} K')

# для 300 см
t_cos5=7000
t_le=300
t_a_300=(1/2)*(t_le*(1-(1-(radius/(radius+h))**2)**(1/2))+t_cos5*(1+(1-(radius/(radius+h))**2)**(1/2)))
print('Для 300 см '+f'{t_a_300} K')
```

Посчитаем моножитель ослабления W  
где E=50 (мкВ/м определили по таблице)

```python
rad=6370*(10**3)    # Радиус земли в метрах
hi=450*(10**3)      # Высота в метрах
```

```python
E=50*(10**(-3))

W=(E*rad)/(3*(10**5))
print ('моножитель ослабления W='+f'{W}')
```
