A2= 1,8

2θ_0,5= 9  шириной диаграммы направленности по половинной мощности  
Вид радиосвязи - Фототелеграфия

β – коэффициент различимости по напряженности поля (5,20);  
А2 – эффективная площадь приемной антенны, м2;  
D1 – коэффициент направленного действия передающей антенны;

W – множитель ослабления;  
T0- температура при которой измеряеться коэф шума (Т0=300 К)
Rз- радиус земли

```python
A2=1.8 # эффективная площадь раскрыва
η2= 0.92
r=4*(10**5)*(10**3) # длинна радиолинии в метрах
Δf= 95*(10**3)  # полоса пропускания в герцвх
T0=300
radius=6370 # радиус земли в километрах
teta=9 # шириной диаграммы направленности по половинной мощности
```

```python
# для 3 см
t_cos1=0
t_a_3=t_cos1
print ('Для 3 см '+f'{t_a_3} K')

# для 15 см
t_cos2=8

t_a_15=t_cos2
print('Для 15 см '+f'{t_a_15} K')

# для 30 см
t_cos3=40
t_a_30=t_cos3
print('Для 30 см '+f'{t_a_30} K')

# для 150 см
t_cos4=1000
t_a_150=t_cos4
print('Для 150 см '+f'{t_a_150} K')

# для 300 см
t_cos5=7000
t_a_300=t_cos5
print('Для 300 см '+f'{t_a_300} K')

```

расчитаем длину радиолинии в метрах

```python
rad=6370*(10**3)    # Радиус земли в метрах
```

Посчитаем моножитель ослабления W

```python
E=50*(10**(-3))
W=(E*rad)/(3*(10**5))
print ('моножитель ослабления W='+f'{W}')
```
