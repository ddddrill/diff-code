расчитаем длину радиолинии для каждого угла

```python
from math import sqrt
print('бери только положительные значения в круглых скобках')

def quadratic_solve(a ,b, c):
    if not all(
        map(
            lambda p: isinstance(p, (int, float)),
            (a, b, c)
        )
    ):
        raise TypeError("Not valid argument type")
    

    if a == 0:
        if b == 0:
            # a и b 0: решения нет
            return None, None
        return -c / b, None
    d = b ** 2 - 4 * a * c
    if d < 0:
        return None, None

    d_root = sqrt(d)
    divider = 2 * a
    x1 = (-b + d_root) / divider
    x2 = (-b - d_root) / divider

    if d == 0:
        x2 = None
    elif x2 > x1:
        x1, x2 = x2, x1

    return x1, x2
print ('для угла 5 гр ')
print(quadratic_solve(1,1110,-11551500))
r_5=2888*(10**3)

print ('для угла 30 гр ')
print(quadratic_solve(1,6370,-11551500))
r_30=1472*(10**3)

print ('для угла 90 гр ')
print(quadratic_solve(1,12740,-11551500))
r_90=850*(10**3)
```
